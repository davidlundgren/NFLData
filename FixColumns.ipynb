```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import datetime
from geopy.distance import vincenty
from matplotlib import ticker
from NFLFunctions import * # custom functions used below
from string import digits
import scipy.optimize as optimization
%matplotlib inline
```

```python
scores = pd.read_csv('/Users/dekedor/NFLData/NFL_Data_DataFrame_V1.csv')
```

```python
scores.head()
```

```python
gamelatlon = [re.split(' +', re.sub('[ ]*\]', '', re.sub('\[[ ]*', '', str(item)))) \
              for item in scores['GameLatLon'].values]
scores['GameLatLon'] = [[0., 0.] if item == ['nan'] else list(map(float, item)) for item in gamelatlon]
awaylatlon = [re.split(' +', re.sub('[ ]*\]', '', re.sub('\[[ ]*', '', str(item)))) \
              for item in scores['AwayLatLon'].values]
scores['AwayLatLon'] = [[0., 0.] if item == ['nan'] else list(map(float, item)) for item in awaylatlon]
```

```python
scores['GameLat'] = np.array([item[0] for item in scores['GameLatLon']])
scores['GameLon'] = np.array([item[1] for item in scores['GameLatLon']])
scores['AwayLat'] = np.array([item[0] for item in scores['AwayLatLon']])
scores['AwayLon'] = np.array([item[1] for item in scores['AwayLatLon']])
```

```python
scores_out = scores.drop(['GameLatLon', 'AwayLatLon', 'Unnamed: 0'], axis=1)
```

```python
scores_out.columns = [code.lower() for code in list(scores_out.keys())]
```

```python
scores_out.to_csv('/Users/dekedor/NFLData/NFL_Data_DataFrame_V2.csv', index=False)
```
